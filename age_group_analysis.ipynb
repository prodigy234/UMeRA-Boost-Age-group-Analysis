```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# If reading from file, use: df = pd.read_excel("UMERA BOOST 3.0.xlsx", sheet_name="UMéRA 3.0")
df = pd.read_excel("UMERA BOOST 2.0 & 3.0.xlsx", sheet_name="UMéRA 3.0")

# Clean column names to remove hidden characters and whitespace
df.columns = df.columns.str.strip().str.replace(r'\u2060', '', regex=True)

# Define the age group column
age_group_column = 'Age Group'

# Handling missing values in Age Group
df[age_group_column] = df[age_group_column].fillna("Unknown")

# Count the number of participants in each age group
age_group_distribution = df[age_group_column].value_counts().sort_index()

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values, palette='viridis', hue=age_group_distribution.index, legend=False)
plt.xlabel("Age Group")
plt.ylabel("Number of Participants")
plt.title("Distribution of Participants Across Age Groups")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Display numerical distribution
print("Age Group Distribution:\n")
print(age_group_distribution)

# Percentage breakdown
age_group_percentage = (age_group_distribution / age_group_distribution.sum()) * 100
print("\nPercentage Breakdown of Age Groups:\n")
print(age_group_percentage.round(2).astype(str) + '%')
```

**Age Group Analysis Report for UMéRA 3.0 Participants**

### **Introduction**
This report provides a detailed analysis of the distribution of participants across different age groups who registered for the UMéRA 3.0 program. Understanding the age demographics of attendees helps in tailoring future training programs, optimizing outreach strategies, and improving participant engagement.

### **Data Overview**
The dataset contains participant registration details, including demographic information such as age group. The key column for this analysis is **“Age Group”**, which classifies participants into predefined age brackets.

### **Findings & Insights**

#### **1. Age Group Distribution**
- The dataset was analyzed to determine the number of participants in each age group.
- The distribution was visualized using a bar chart, showing the frequency of each category.

#### **2. Key Observations**
- The highest number of participants belong to the **18-24** age group, indicating that students and fresh graduates are the primary audience.
- The **25-34** age group also has a significant presence, suggesting strong interest from young professionals and early-career individuals.
- The **35-44** and **45+** groups have lower representation, indicating fewer mid-career and senior professionals enrolling in the program.
- A small percentage of participants had missing or unclassified age data, which could indicate errors in data collection or voluntary omissions.

#### **3. Percentage Breakdown**

- **35-44 years:** 10.31% (moderate engagement, likely career professionals)
- **45+ years:** 2.55% (lowest participation, indicating less demand from senior professionals)
